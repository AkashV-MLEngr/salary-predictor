```python
#Histograms of all numerical columns 

num_col = [col for col in data.columns if data[col].dtype != 'object']

fig, axs = plt.subplots(1, 3, figsize=(25, 8))
axs = axs.flatten()

for i, col in enumerate(num_col):
    sns.histplot(data=data, x=col, ax=axs[i], kde=True)
    axs[i].set_title(f'{col}')

plt.tight_layout()
plt.show()
```

```python
# EDA - AVG [Salary, Age, Experience] of Different Education, Location, Gender, Job_Title

def avg_num(num_col):
    
    fig, axs = plt.subplots(2, 2, figsize=(16, 8))
    axs = axs.flatten()

    for i, col in enumerate(cat_col):
        df=data.groupby(col)[num_col].mean().reset_index().sort_values(num_col, ascending=False)

        sns.barplot(
            data=df, x=col, y=num_col, ax=axs[i]
        )
        axs[i].set_title(f'Avg {num_col} of Different Category Under {col}')

    plt.tight_layout()
    plt.show()

# avg_num('Experience')
# avg_num('Age')
avg_num('Salary')
```

```python
# AVG [Salary, Age, Experience] of Different [Job_Title, Education, Location, gender]

def avg_job_salary(col, col2, num):
    num_cols=2
    num_rows=math.ceil(len(data[col].unique())/num_cols)
    
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(16, 8))
    axs = axs.flatten()

    for i, variable in enumerate(data[col].unique()):
        df = data.loc[data[col]==variable]

        df = df.groupby(col2)[num].mean().sort_values(ascending=False).reset_index()

        sns.barplot(data=df, x=col2, y=num, ax=axs[i])
        axs[i].set_title(f'AVG {num} of Each {col2} - {variable}')

    plt.tight_layout()
    plt.show()
    
# avg_job_salary('Gender', 'Location', 'Salary')
# avg_job_salary('Location', 'Education', Salary')
avg_job_salary('Education', 'Job_Title', 'Experience')
# avg_job_salary('Education', 'Location', 'Salary')
```

Salary Prediction (Regression)

```python
#Use Label Encoder to tranfrom num_cols into numbers
from sklearn.preprocessing import LabelEncoder

label_encode = LabelEncoder()

for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = label_encode.fit_transform(data[col])
```

```python
#Divide the data to X and y
X=data.drop('Salary', axis=1)
y=data.loc[:, 'Salary']

#Scale the data
from sklearn.preprocessing import StandardScaler

std_scale = StandardScaler()

X = pd.DataFrame(std_scale.fit_transform(X), columns=[X.columns])
```

```python
#Select columns with most significance
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

model = SelectFromModel(Lasso(0.005))
model.fit(X, y)

#Every Columns need to be used this time for prediction
model.get_support()
```

```python
#Split the data to Train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Create Function that run prediction and return results
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score


def predict(ml_model):
    model = ml_model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print(f'Predictions: {pred[:20]} ...')
    print('\n')
    print(f'Mean Squared Error: {np.sqrt(mean_squared_error(y_test, pred))}')
    print('\n')
    print(f'mean_absolute_error: {mean_absolute_error(y_test, pred)}')
    print('\n')
    print(f"Accuracy Percentage: {(r2_score(y_test, pred))*100}")
    # print(f"Accuracy Score: {(accuracy_score(y_test, pred))*100}")

    
    #Cross Validated Results
    def display_scores(scores):
        print("==================================")
        print("Scores: {}".format(scores))
        print("==================================")
        print("Mean Score: {}".format(scores.mean()))
        print("==================================")
        print("Standard Deviation of Scores: {}".format(scores.std()))
        print("==================================")
        

        return None

    from sklearn.model_selection import cross_val_score

    scores = cross_val_score(ml_model, X_train, y_train, 
                            scoring='neg_mean_squared_error', cv=10)
    
    tree_rmse_in_sample_estimates = np.sqrt(-scores)
    print('Cross Validated Score:')
    print('\n')
    display_scores(tree_rmse_in_sample_estimates)

    #Plot the results
    df=pd.DataFrame({'Prediction':pred, 'y_test':y_test})
    sns.regplot(data=df, x='Prediction', y='y_test', line_kws={"color": "blue"})
    plt.show()
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
#predict(forest)
forest = forest.fit(X_train, y_train)
pred = forest.predict(X_test)
print(f'Predictions: {pred[:20]} ...')
print('\n')
print(f'Mean Squared Error: {np.sqrt(mean_squared_error(y_test, pred))}')
print('\n')
print(f'mean_absolute_error: {mean_absolute_error(y_test, pred)}')
print('\n')
print(f"Accuracy Percentage: {(r2_score(y_test, pred))*100}")
```
